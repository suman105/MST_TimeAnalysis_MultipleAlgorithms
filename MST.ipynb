```python
import sys
import heapq
import timeit
import random
import matplotlib.pyplot as plt
import networkx as nx
from collections import defaultdict

def add_edge(graph, u, v, w):
    graph[u].append((v, w))
    graph[v].append((u, w))  # Assuming the graph is undirected

def prim_mst(graph, vertices,i):
    key = [sys.maxsize] * vertices
    parent = [None] * vertices
    mst_set = [False] * vertices
    min_heap = [(0, 0)]  # (key, vertex)

    key[0] = 0  # Start from the first vertex
    total_distance = 0

    while min_heap:
        current_key, u = heapq.heappop(min_heap)
        if mst_set[u]:
            continue

        mst_set[u] = True
        total_distance += current_key

        for v, weight in graph[u]:
            if not mst_set[v] and key[v] > weight:
                key[v] = weight
                parent[v] = u
                heapq.heappush(min_heap, (weight, v))

    mst_edges = [(parent[i], i) for i in range(1, vertices)]

    if i==0:
        return
    return mst_edges, total_distance

# Function to visualize the graph and MST
def visualize_graph(graph, mst_edges, distance):
    G = nx.Graph()

    for u in graph:
        for v, weight in graph[u]:
            G.add_edge(u, v, weight=weight)

    pos = nx.spring_layout(G)  # You can change the layout algorithm as needed
    plt.figure(figsize=(7,5))
    nx.draw(G, pos, with_labels=True, font_weight='bold', node_size=500, node_color='forestgreen', font_color='black')

    # Highlight MST edges
    nx.draw_networkx_edges(G, pos, edgelist=mst_edges, edge_color='red', width=2)

    edge_labels = {(u, v): d['weight'] for u, v, d in G.edges(data=True)}
    nx.draw_networkx_edge_labels(G, pos, edge_labels=edge_labels)

    #plt.text(0.2, 0.05, f'MST Distance: {distance:.2f}', horizontalalignment='center', verticalalignment='center', transform=plt.gca().transAxes)
    plt.title("g")
    plt.show()

# # Dense Input Case
# def dense_graph(num_vertices):
#     graph = defaultdict(list)

#     for i in range(num_vertices):
#         for j in range(i + 1, num_vertices):
#             weight = random.randint(5, 20)  # Random weight between 5 and 20
#             graph[i].append((j, weight))
#             graph[j].append((i, weight))  # Assuming the graph is undirected

#     return graph

# # Sparse Graph Input
# def sparse_graph(num_vertices, avg_degree):
#     graph = defaultdict(list)

#     for i in range(num_vertices):
#         neighbors = random.sample(range(i + 1, num_vertices), min(avg_degree, num_vertices - i - 1))
#         for j in neighbors:
#             weight = random.randint(5, 20)  # Random weight between 5 and 20
#             graph[i].append((j, weight))
#             graph[j].append((i, weight))  # Assuming the graph is undirected

#     return graph

# # Average Graph Input
# def average_graph(num_vertices):
#     graph = defaultdict(list)

#     for i in range(num_vertices):
#         # Connect to about half of the other nodes
#         num_connections = num_vertices // 2
#         neighbors = random.sample([j for j in range(num_vertices) if j != i], num_connections)

#         for j in neighbors:
#             weight = random.randint(1, 20)  # Random weight between 1 and 20
#             graph[i].append((j, weight))
#             graph[j].append((i, weight))  # Assuming the graph is undirected

#     return graph

if __name__ == "__main__":
    vertices = 25
    v=75
    ve = 7
    gd = defaultdict(list)
    gs = defaultdict(list)
    ga = defaultdict(list)

    d = defaultdict(list)
    s = defaultdict(list)
    a = defaultdict(list)

    gdd = defaultdict(list)
    gss = defaultdict(list)
    gaa = defaultdict(list)

    for i in range(ve):
        for edge in d_input[i]:
            add_edge(gd, i, edge[0], edge[1])

    for i in range(ve):
        for edge in s_input[i]:
            add_edge(gs, i, edge[0], edge[1])

    for i in range(ve):
        for edge in a_input[i]:
            add_edge(ga, i, edge[0], edge[1])

    for i in range(vertices):
        for edge in dense_graph_input[i]:
            add_edge(gd, i, edge[0], edge[1])

    for i in range(vertices):
        for edge in sparse_graph_input[i]:
            add_edge(gs, i, edge[0], edge[1])

    for i in range(vertices):
        for edge in average_graph_input[i]:
            add_edge(ga, i, edge[0], edge[1])

    for i in range(v):
        for edge in dense_input[i]:
            add_edge(gdd, i, edge[0], edge[1])

    for i in range(v):
        for edge in sparse_input[i]:
            add_edge(gss, i, edge[0], edge[1])

    for i in range(v):
        for edge in avgerage_input[i]:
            add_edge(gaa, i, edge[0], edge[1])

    num_trials=1000

    print("Results from Prim's Algorithm:")

    time_d = timeit.timeit(lambda: prim_mst(d, 7,0), number=num_trials)
    time_s = timeit.timeit(lambda: prim_mst(s, 7,0), number=num_trials)
    time_a = timeit.timeit(lambda: prim_mst(a, 7,0), number=num_trials)

    time_ddense = timeit.timeit(lambda: prim_mst(gdd, 75,0), number=num_trials)
    time_ssparse = timeit.timeit(lambda: prim_mst(gss, 75,0), number=num_trials)
    time_aavg = timeit.timeit(lambda: prim_mst(gaa, 75,0), number=num_trials)

    time_dense = timeit.timeit(lambda: prim_mst(gd, 25,0), number=num_trials)
    time_sparse = timeit.timeit(lambda: prim_mst(gs, 25,0), number=num_trials)
    time_avg = timeit.timeit(lambda: prim_mst(ga, 25,0), number=num_trials)

    mst_edges_dense, total_distance_dense = prim_mst(gd, 25,1)
    mst_edges_sparse, total_distance_sparse = prim_mst(gs, 25,1)
    mst_edges_average, total_distance_average = prim_mst(ga, 25,1)


    visualize_graph(d_input, mst_edges_dense,total_distance_dense)
    visualize_graph(s_input, mst_edges_sparse, total_distance_sparse)
    visualize_graph(a_input, mst_edges_average, total_distance_average)



    print("MST Distance from Prim's using Dense Graph Input:", total_distance_dense)
    print("MST Distance from Prim's using Sparse Graph Input:", total_distance_sparse)
    print("MST Distance from Prim's using Average Graph Input:", total_distance_average)

    print("Execution Time from Prim's Dense Graph Input:", time_dense)
    print("Execution Time from Prim's Sparse Graph Input:", time_sparse)
    print("Execution Time from Prim's Average Graph Input:", time_avg)
    print("\n")
    print("Execution Time from Prim's Dense Graph Input:", time_ddense)
    print("Execution Time from Prim's Sparse Graph Input:", time_ssparse)
    print("Execution Time from Prim's Average Graph Input:", time_aavg)
    print("\n")
    print("Execution Time from Prim's Dense Graph Input:", time_d)
    print("Execution Time from Prim's Sparse Graph Input:", time_s)
    print("Execution Time from Prim's Average Graph Input:", time_a)


    graph_types = ['Dense', 'Sparse', 'Average']
    execution_times = [1.322775293999257, 0.3855095290000463, 0.3468741839997165]
    execution = [1.322775293999257, 0.16059233999931166, 1.3176021359995502]
    e = [0.0036689060007120133, 0.003643291000116733, 0.0036198349998812773]


    plt.bar(graph_types, e, color=['blue', 'green', 'orange'])
    plt.xlabel('Graph Type')
    plt.ylabel('Execution Time (s)')
    plt.title('Prim\'s Algorithm Performance for 7 vertices')
    plt.show()

    plt.bar(graph_types, execution_times, color=['blue', 'green', 'orange'])
    plt.xlabel('Graph Type')
    plt.ylabel('Execution Time (s)')
    plt.title('Prim\'s Algorithm Performance for 25 Vertices')
    plt.show()

    plt.bar(graph_types, execution, color=['blue', 'green', 'orange'])
    plt.xlabel('Graph Type')
    plt.ylabel('Execution Time (s)')
    plt.title('Prim\'s Algorithm Performance for 75 vertices')
    plt.show()
```
